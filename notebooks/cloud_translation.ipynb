```python
from google.cloud import translate
```

```python
def translate_text(text="Hello, world!", project_id="my-project-syaco"):

    client = translate.TranslationServiceClient()
    location = "global"
    parent = f"projects/{project_id}/locations/{location}"

    response = client.translate_text(
        request={
            "parent": parent,
            "contents": [text],
            "mime_type": "text/plain",
            "source_language_code": "en-US",
            "target_language_code": "es",
        }
    )

    for translation in response.translations:
        print("Translated text: {}".format(translation.translated_text))

```

```python
from glob import glob
import os

#translation libraries
!pip -q install googletrans==4.0.0rc1
from googletrans import Translator
import subprocess
import json
```

```python
# Googletrans translate en to japanese, removing blank lines first

# Function to remove blank lines from a file
def remove_blank_lines(input_file):
    with open(input_file, "r", encoding="utf-8") as file:
        lines = file.readlines()
    # Remove blank lines and leading/trailing whitespace
    lines = [line.strip() for line in lines if line.strip()]
    return lines

# Function to translate text from English to Japanese
def translate_to_japanese(etext):
    translator = Translator()
    translated = translator.translate(etext,dest="ja")
    #translated = translator.translate(text, src="en", dest="ja")
    return translated


# Remove blank lines from the input file
lines = remove_blank_lines(input_file)

# Translate each line from English to Japanese
translated_lines = [translate_to_japanese(line) for line in lines]

# Write the translated text to the output file
with open(output_file, "w", encoding="utf-8") as file:
    for line in translated_lines:
        file.write(line.text + "\n")

print("Translation completed. Output saved to:", output_file)
```
