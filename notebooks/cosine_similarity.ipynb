```python
#set up paths
# Define the path to your text files
text_path = "data"
input_file= text_path + "/Griffis_Diary_corrected.txt"
input_file2= text_path + "/Griffis_Diary_transcribed.txt"
```

```python
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import os
```

```python
# Ensure that the files exist before proceeding
if not os.path.isfile(input_file) or not os.path.isfile(input_file2):
    raise FileNotFoundError('One of the input files does not exist.')
with open(input_file, 'r', encoding='utf-8') as file:
    doc1 = file.read()

with open(input_file2, 'r', encoding='utf-8') as file:
    doc2 = file.read()

# Initialize the vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = tfidf_vectorizer.fit_transform([doc1, doc2])

# Calculate Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
# Print out the result
print(f'Cosine Similarity between', input_file, 'and', input_file2)

print(f'{cosine_sim[0][0]*100:.2f}%')
```
